```python
#Imports
import sys
import os
import numpy as np
import matplotlib.pyplot as plt
import random
import time
import tensorflow
import keras
from tqdm.auto import tqdm

from tensorflow.keras.callbacks import EarlyStopping

print(keras.__version__)
print(tensorflow.__version__)
sys.path.insert(1, r"..\src")
import utils
from nn_architectures import rf_nn
from nn_architectures import rf_nn_2
```

```python
# Load data    
folder = r'\Supplementary_5' #or Supplementary 5
assert(folder=='\Main_3' or folder=='\Supplementary_5'),"Expected the folder to be r'\Main_3' or r'\Supplementary_5'!"
train_images = np.load(r'..\data\02_data_RF{}\training_data\dummy_train_images.npy'.format(folder))
train_histograms = np.load(r'..\data\02_data_RF{}\training_data\dummy_train_histograms.npy'.format(folder))
test_images = np.load(r'..\data\02_data_RF{}\test_data\test_images.npy'.format(folder))
test_histograms = np.load(r'..\data\02_data_RF{}\test_data\test_histograms.npy'.format(folder))
time_series = np.load(r'..\data\02_data_RF\time_series.npy')

# Validation: we aren't validating the models at this stage, we are just tracking the progress of the neural network, which is 
# why we set the validation set to be the test set. When validating previously, we set the first 1000 elements of the training set
# as validation data (and cropped the training set accordingly)
val_histograms=test_histograms.copy()
val_images=test_images.copy()

# Define parameters of the simulation
if folder == r'\Main_3':
    max_depth = 4.392669921875 # this is the maximum distance in the room, from the camera: 4.4 m
elif folder ==r'\Supplementary_5':
    max_depth = 4.6
```

```python
# Crop the datasets in order to select the appropriate subset. In order of increasing number of included scattering events, 
# crop the histograms at: 86, 142, 198, 256
if folder==r'\Main_3':
    test_histograms[:,86::] = 0 # I select subset 1 here
    train_histograms[:,86::] = 0
```

```python
#%% Plot an image-histogram pair, just to check correct loading
frame = 102

plt.close('all')
fig=plt.figure(figsize=(10,3))

plt.subplot(121)
plt.plot(time_series*1e9,test_histograms[frame])
plt.title('(Original) histogram')
plt.xlabel('Time (ns)')

plt.subplot(122)
image = test_images[frame]
plt.imshow(image*max_depth, cmap='viridis_r',vmin=0,vmax=max_depth)
cbar=plt.colorbar(orientation='vertical')
cbar.set_label('Depth (m)')
plt.title('Depth map')

plt.show()
```

```python
if 'model' in locals(): del model
keras.backend.clear_session()

def train(epochs, feats, batch, kern_int_e, kern_int_d, kern_reg, kse, ksd, val):
    
    if folder=='\Main_3':
        model = rf_nn(feats, kern_int_e, kern_int_d, kern_reg, kse, ksd)
    elif folder=='\Supplementary_5':
        model = rf_nn_2(feats, kern_int_e, kern_int_d, kern_reg, kse, ksd)
    adam = tensorflow.keras.optimizers.Adam(learning_rate=0.0001, beta_1=0.9, beta_2=0.999)
    model.compile(loss='mse', optimizer=adam, metrics=['accuracy'])

    #enable callbacks for validation
    if val==True:
      my_callbacks = [EarlyStopping(monitor='val_loss',
                                    patience=20)]
    #disable callbacks for training final model
    if val==False:
      my_callbacks = []
                        

    start=time.time()
    history = model.fit(train_histograms, train_images, epochs=epochs, batch_size=batch, 
                        validation_data=(val_histograms, val_images), verbose=1, 
                        callbacks=my_callbacks)
    end=time.time()
    
    fig=plt.figure(figsize=(5,3))
    plt.title('Train with {} datapoints'.format(len(train_images)))
    plt.plot(history.history['loss'],label='training loss')
    plt.plot(history.history['val_loss'],label='validation loss')
    plt.xlabel('Epoch')
    plt.ylabel('MSE')
    plt.yscale('log')
    plt.legend()    
    plt.savefig(r'..\data\02_data_RF\models\dummy_training_progression.h5')
    plt.show()
    
    print('comp. time =',end-start)
    model.save(r'..\data\02_data_RF\models\dummy_model.h5')
    return model
```

```python
# Fit model with GPU
utils.reset_keras()
model = train(epochs=500, feats=64, batch=64, kern_int_e='he_normal', kern_int_d='he_normal', kse=7, ksd=5, kern_reg=None, val=False)
```
