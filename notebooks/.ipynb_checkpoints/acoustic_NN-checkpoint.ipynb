```python
#Imports
import sys
import os
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import random
import time
import tensorflow
import keras
from tqdm.auto import tqdm

from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.regularizers import l2, l1, l1_l2

print(keras.__version__)
print(tensorflow.__version__)
sys.path.insert(1, r"..\src")
import utils
from nn_architectures import acoustic_nn
```

```python
# Load data    
test_images = np.load(r'..\data\03_data_acoustic\test_data\test_images.npy')
test_histograms = np.load(r'..\data\03_data_acoustic\test_data\test_histograms.npy')
train_images = test_images[0:100,:,:].copy() #dummy training images; in our actual research, we used a separate training set of course
train_histograms = test_histograms[:,0:100,:,:].copy()
time_series = np.load(r'..\data\03_data_acoustic\time_series.npy')

# Validation: we aren't validating the models at this stage, we are just tracking the progress of the neural network, which is 
# why we set the validation set to be the test set. When validating previously, we set the first 1000 elements of the training set
# as validation data (and cropped the training set accordingly)
val_histograms=test_histograms.copy()
val_images=test_images.copy()



# Define parameters of the simulation
max_depth = 5 #from the experiment, this is the maximum distance in the room, from the camera: 5 m
```

```python
#%% Plot an image-histogram pair, just to check correct loading
subset=2
frame = 15

plt.close('all')
fig=plt.figure(figsize=(10,3))

plt.subplot(121)
plt.plot(time_series,test_histograms[subset,frame])
plt.title('(Original) histogram')
plt.xlabel('Time (ms)')

plt.subplot(122)
image = test_images[subset,frame].reshape(64,64)
plt.imshow(image*max_depth, cmap='viridis',vmin=0,vmax=max_depth)
cbar=plt.colorbar(orientation='vertical',mappable=plt.cm.ScalarMappable(cmap='viridis_r',norm=matplotlib.colors.Normalize(vmin=0, vmax=max_depth)))
cbar.set_label('Depth (m)')
plt.title('Depth map')

plt.show()
```

```python
def train(epochs, subset, feats, batch, kern_int_e, kern_int_d, kern_reg, kse, ksd, val):
    
    model = acoustic_nn(feats, kern_int_e, kern_int_d, kern_reg, kse, ksd)
    adam = tensorflow.keras.optimizers.Adam(learning_rate=0.0001, beta_1=0.9, beta_2=0.999)
    model.compile(loss='mse', optimizer=adam, metrics=['accuracy'])

    #enable callbacks for validation
    if val==True:
      my_callbacks = [EarlyStopping(monitor='val_loss',
                                    patience=20)]
    #disable callbacks for training final model
    if val==False:
      my_callbacks = []
                        

    start=time.time()
    history = model.fit(train_histograms[subset], train_images[subset], epochs=epochs, batch_size=batch, 
                        validation_data=(val_histograms[subset], val_images[subset]), verbose=1, 
                        callbacks=my_callbacks)
    end=time.time()
    
    fig=plt.figure(figsize=(5,3))
    plt.title('Train with {} datapoints'.format(len(train_images)))
    plt.plot(history.history['loss'],label='training loss')
    plt.plot(history.history['val_loss'],label='validation loss')
    plt.xlabel('Epoch')
    plt.ylabel('MSE')
    plt.yscale('log')
    plt.legend()    
    plt.savefig(r'..\data\03_data_acoustic\models\dummy_training_progression.h5')
    plt.show()
    
    print('comp. time =',end-start)
    model.save(r'..\data\03_data_acoustic\models\dummy_model.h5')
```

```python
# Dummy training
utils.reset_keras()
train(epochs=5, subset=2, feats=64, batch=64, kern_int_e='he_normal', kern_int_d='he_normal', kse=7, ksd=5, kern_reg=None, val=False)
```
