```python
#Imports

import sys
import os
import numpy as np
import matplotlib.pyplot as plt
import random
import time
from matplotlib import gridspec
from skimage.metrics import structural_similarity
from sklearn.metrics import mean_squared_error
from tqdm.auto import tqdm
import matplotlib
import tensorflow as tf

sys.path.insert(1, r"..\src")
import utils
```

```python
max_depth = 5

#Load data
test_images = np.load(r"..\data\03_data_acoustic\test_data\test_images.npy")
test_histograms = np.load(r"..\data\03_data_acoustic\test_data\test_histograms.npy")
predictions = np.zeros((4,500,64,64,1))

#Reconstruct 3D images from the corresponding temporal histograms using the respective CNNs
for i,j in enumerate(['24','48','72','full']):
    print('Subset',i,'model',j)
    model = tf.keras.models.load_model(r"..\data\03_data_acoustic\models\acoustics_model_{}.h5".format(j))
    predictions[i] = model.predict(test_histograms[i])
predictions = predictions.reshape(((4,500,64,64)))
predictions = predictions.astype('float32')
    
#Additional data    
time_series = np.load(r"..\data\03_data_acoustic\time_series.npy")
background_mask = np.load(r"..\data\03_data_acoustic\test_data\background_mask.npy")
```

```python
# Plot examples for visualisation
n = 2 # subset
co = 200 # image index

plt.figure(figsize=(10,3),dpi=72)
plt.subplot(122)
plt.title('Ground truth',fontname='Arial',fontsize=12)
plt.imshow(test_images[n,co]*max_depth,cmap='viridis',vmin=0,vmax=np.ceil(max_depth))
cbar=plt.colorbar(orientation='vertical',mappable=plt.cm.ScalarMappable(cmap='viridis_r',norm=matplotlib.colors.Normalize(vmin=0, vmax=max_depth)))
cbar.set_label('Depth (m)')
plt.subplot(121)
plt.title('Reconstruction',fontname='Arial',fontsize=12)
plt.imshow(predictions[n,co]*max_depth,cmap='viridis',vmin=0,vmax=np.ceil(max_depth))
cbar=plt.colorbar(orientation='vertical',mappable=plt.cm.ScalarMappable(cmap='viridis_r',norm=matplotlib.colors.Normalize(vmin=0, vmax=max_depth)))
cbar.set_label('Depth (m)')
plt.show()
fig=plt.figure(figsize=(5,1.5),dpi=72)
plt.tick_params(direction='in')
plt.plot(time_series,test_histograms[n,co])
plt.xticks(fontsize=12,fontname='Arial')
plt.yticks(fontsize=12,fontname='Arial')
plt.xlabel('Time (ns)',fontsize=12,fontname='Arial')
plt.ylabel('Counts (a.u.)',fontsize=12,fontname='Arial')
plt.show()
```

### Metrics; use k=0.2 for acoustic data

```python
# View IOU
fmse, iou = utils.fmse_iou(predictions[2,200]*max_depth,test_images[2,200]*max_depth, mask=background_mask,k=0.2,acoustic=True,show=True) # returns the mean squared error over the union of foregrounds,
                                                                                                  # and the iou  
```

```python
# Obtain performance metrics
mse_list=np.zeros((4,865))
fmse_list=np.zeros((4,865))
ssim_list=np.zeros((4,865))
iou_list=np.zeros((4,865))
for i in tqdm(range(4)):
    for co in range(test_histograms.shape[1]):
        mse_list[i,co]=(mean_squared_error(test_images[i,co]*max_depth,predictions[i,co]*max_depth))
        ssim_list[i,co]=structural_similarity(test_images[i,co]*max_depth,predictions[i,co]*max_depth)
        fmse_list[i,co],iou_list[i,co]=utils.fmse_iou(predictions[i,co]*max_depth,test_images[i,co]*max_depth,mask=background_mask,k=0.2)    
        
mean_mse = np.mean(mse_list,axis=1)
mean_ssim = np.mean(ssim_list,axis=1)
mean_iou = np.mean(iou_list,axis=1)
mean_fmse = np.mean(fmse_list,axis=1)

fig = plt.figure(figsize=(5,2),dpi=72)
ax1 = fig.add_subplot(111)
plt.title('Quality of Predictions')
ax1.set_xticks((range(1,5)))
ax1.set_xticklabels(['t<25.0','t<41.5','t<58.1','full'])
plt.xlabel('Subset')
plt.grid(b=True, axis='y')

#MSE
color = 'tab:orange'
ax1.set_ylabel('MSE',color=color) 
ax1.tick_params(direction='in')
mse = ax1.plot(range(1,5),mean_mse, color=color, label='MSE',marker='o')
#f_mse = ax1.plot(range(1,5),mean_fmse,color='orange',label='Foreground MSE')
ax1.tick_params(axis='y', labelcolor=color)
#ax1.set_ylim([0,0.005])
#ax1.legend()

##SSIM
ax2 = ax1.twinx()

color = 'tab:blue'
ax2.set_ylabel('IOU', color=color) 
ax2.tick_params(direction='in')
iou = ax2.plot(range(1,5),mean_iou, color=color, label='IOU',marker='o')
#ssim= ax2.plot(range(1,5),mean_ssim, color='black', label='SSIM')
#ax2.set_ylim([0.85,0.95])
ax2.tick_params(axis='y', labelcolor=color)

# lns = mse+iou+f_mse+ssim
# labs = [l.get_label() for l in lns]
# ax1.legend(lns, labs, loc=5)

plt.show()
```
