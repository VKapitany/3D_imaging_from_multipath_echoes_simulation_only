```python
#Imports

import sys
import os
import numpy as np
import matplotlib.pyplot as plt
import random
import time
from matplotlib import gridspec
from skimage.metrics import structural_similarity
from sklearn.metrics import mean_squared_error
from tqdm.auto import tqdm
import tensorflow as tf

sys.path.insert(1, r"..\src")
import utils
```

```python
max_depth = 4.392669921875

#Load data
test_images = np.load(r"..\data\02_data_RF\Main_3\test_data\test_images.npy")
test_histograms_0 = np.load(r"..\data\02_data_RF\Main_3\test_data\test_histograms.npy")

test_histograms = np.zeros((4,test_histograms_0.shape[0],test_histograms_0.shape[1],test_histograms_0.shape[2]))
test_histograms[0,:,:,:] = test_histograms_0.copy()
test_histograms[0,:,86:256,:]=0
test_histograms[1,:,:,:] = test_histograms_0.copy()
test_histograms[1,:,142:256,:]=0
test_histograms[2,:,:,:] = test_histograms_0.copy()
test_histograms[2,:,198:256,:]=0
test_histograms[3,:,:,:] = test_histograms_0.copy()

#Reconstruct 3D images from RF histograms
predictions = np.zeros((4,865,60,80,1))
for i,j in enumerate(['25','41','58','full']):
    print('Subset',i,'model',j)
    model = tf.keras.models.load_model(r"..\data\02_data_RF\Main_3\models\cae_subset{}.h5".format(j))
    predictions[i] = model.predict(test_histograms[i])
predictions = predictions.reshape((4,865,60,80))
predictions = predictions.astype('float64')

#Load additional data
time_series = np.load(r"..\data\02_data_RF\time_series.npy")
background_mask = np.load(r"..\data\02_data_RF\Main_3\test_data\background_mask.npy")
```

```python
# Plot examples for visualisation
n = 2 # subset
co = 500 # image index

plt.figure(figsize=(10,3),dpi=72)
plt.subplot(122)
plt.title('Ground truth',fontname='Arial',fontsize=12)
plt.imshow(test_images[co,:,:]*max_depth,cmap='viridis_r',vmin=0,vmax=np.ceil(max_depth))
cbar=plt.colorbar()
cbar.set_label('Depth (m)',rotation=270,labelpad=15,fontsize=12,fontname='Arial')
cbar.ax.tick_params(direction='in')
plt.subplot(121)
plt.title('Reconstruction',fontname='Arial',fontsize=12)
plt.imshow(predictions[n,co,:,:]*max_depth,cmap='viridis_r',vmin=0,vmax=np.ceil(max_depth))
cbar=plt.colorbar()
cbar.set_label('Depth (m)',rotation=270,labelpad=15,fontsize=12,fontname='Arial')
cbar.ax.tick_params(direction='in')
plt.show()
fig=plt.figure(figsize=(5,1.5),dpi=72)
plt.tick_params(direction='in')
plt.plot(time_series,test_histograms[n,co])
plt.xticks(fontsize=12,fontname='Arial')
plt.yticks(fontsize=12,fontname='Arial')
plt.xlabel('Time (ns)',fontsize=12,fontname='Arial')
plt.ylabel('Counts (a.u.)',fontsize=12,fontname='Arial')
plt.show()
```

### Metrics; use k=0.2 for radar data

```python
# View IOU
fmse, iou = utils.fmse_iou(predictions[2,500]*max_depth,test_images[500]*max_depth, mask=background_mask,k=0.2,show=True) # returns the mean squared error over the union of foregrounds,
                                                                                                  # and the iou  
```

```python
# Obtain performance metrics
mse_list=np.zeros((4,865))
fmse_list=np.zeros((4,865))
ssim_list=np.zeros((4,865))
iou_list=np.zeros((4,865))
for i in tqdm(range(4)):
    for co in range(test_histograms.shape[1]):
        mse_list[i,co]=(mean_squared_error(test_images[co]*max_depth,predictions[i,co]*max_depth))
        ssim_list[i,co]=structural_similarity(test_images[co]*max_depth,predictions[i,co]*max_depth)
        fmse_list[i,co],iou_list[i,co]=utils.fmse_iou(predictions[i,co]*max_depth,test_images[co]*max_depth,mask=background_mask,k=0.2)    
        
mean_mse = np.mean(mse_list,axis=1)
mean_ssim = np.mean(ssim_list,axis=1)
mean_iou = np.mean(iou_list,axis=1)
mean_fmse = np.mean(fmse_list,axis=1)

fig = plt.figure(figsize=(5,2),dpi=72)
ax1 = fig.add_subplot(111)
plt.title('Quality of Predictions')
ax1.set_xticks((range(1,5)))
ax1.set_xticklabels(['t<25.0','t<41.5','t<58.1','full'])
plt.xlabel('Subset')
plt.grid(b=True, axis='y')

#MSE
color = 'tab:orange'
ax1.set_ylabel('MSE',color=color) 
ax1.tick_params(direction='in')
mse = ax1.plot(range(1,5),mean_mse, color=color, label='MSE',marker='o')
ax1.tick_params(axis='y', labelcolor=color)

##IOU
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('IOU', color=color) 
ax2.tick_params(direction='in')
iou = ax2.plot(range(1,5),mean_iou, color=color, label='IOU',marker='o')
ax2.tick_params(axis='y', labelcolor=color)

plt.show()
```
