### Model architecture

The model architecture of choice is an hourglass shaped, fully convolutional neural network.

```python
if 'model' in locals(): del model
keras.backend.clear_session()

def train(nscatt, train_inputs, train_labels, val_inputs, val_labels, epochs, feats, batch, kern_int_e, kern_int_d, kern_reg, kse, ksd, val):
    
    model = simulation_nn(feats, kern_int_e, kern_int_d, kern_reg, kse, ksd)
    adam = tf.keras.optimizers.Adam(learning_rate=5e-5, beta_1=0.9, beta_2=0.999)
    model.compile(loss='mse',optimizer=adam, metrics=['accuracy'])

    #enable callbacks for validation
    if val==True:
      my_callbacks = [EarlyStopping(monitor='val_loss',
                                    patience=10)]
    #disable callbacks for training final model
    if val==False:
      my_callbacks = []
                        
    
    start=time.time()
    history = model.fit(train_inputs, train_labels, epochs=epochs, batch_size=batch, 
                        validation_data=(val_inputs, val_labels), verbose=1, 
                        callbacks=my_callbacks)
    end=time.time()
     
    print('comp. time =',end-start)
    return model, history.history['loss'], history.history['val_loss']
```

### Training pipeline

We train the models in two steps:

- For each number of max. scattering events, we select the first 200 of the 2000 training data for validation, and allocate the the rest for training. We train 3 models using an EarlyStopping callback, to limit the number of epochs the models train for based on the performance on the validation data. We calculate the integer mean of the number of epochs at which the minimal validation loss was achieved. The number of epochs is the only hyperparameter we properly searched; the other parameters were inherited from other, related projects and were not optimised.

- Having validated the the number of epochs, we train 10 examples of neural networks to avoid specificity of our results on a certain NN starting configuration. For this, we assign the full training set for training. From each neural network, we predict outputs for the test set inputs, and save them.

### Dummy training

```python
epochs=5
n=1 #Index of dataset, where n = max number of scattering events - 1
model, train_hist, val_hist = train(nscatt = n+1,
                            train_inputs = train_histograms[n],
                            train_labels = train_images[n],
                            val_inputs = test_histograms[n],
                            val_labels = test_images[n],
                            epochs=epochs,
                            feats=8,
                            batch=100,
                            kern_int_e='he_normal',
                            kern_int_d='he_normal',
                            kse=7,
                            ksd=5,
                            kern_reg=None,
                            val=False)
```
